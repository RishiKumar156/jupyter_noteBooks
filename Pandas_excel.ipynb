```python
import pandas as pd 
df = pd.read_csv("BooksDataSet.csv", index_col=False)
df['Authors'] = df['Authors'].str.replace("By", '')
df['Price'] = df['Price'].str.replace("Price Starting at", '')
# Clean up the 'Publish Date' column using regex
days_to_remove = ["Friday", "Saturday", "Sunday", "Monday", "Tuesday", "Wednesday", "Thursday"]
days_pattern = '|'.join(days_to_remove)
"""so here we are going to split the data with months and years """
df['Publish Date'] = df['Publish Date'].str.replace(f'({days_pattern}),', '', regex=True)
df['Publish Date'] = pd.to_datetime(df['Publish Date'],errors='coerce')
df['moths'] = df['Publish Date'].dt.month
df['years'] = df['Publish Date'].dt.year
df = df.drop_duplicates()
df
```
